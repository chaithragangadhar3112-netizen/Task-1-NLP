```python
#Task 2: Preprocessing Function
import re

def preprocess_text(text):
    # 1. Handle empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""

    # 2. Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|\S+@\S+", "", text)

    # 3. Remove numbers
    text = re.sub(r"\d+", "", text)

    # 4. Convert to lowercase
    text = text.lower()

    # 5. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)

    # 6. Remove special characters (keep only alphabets & spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Tokenization
    tokens = text.split()

    # 9. Remove short words (≤2 except 'no', 'not')
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)

    # 10. Join back to sentence
    clean_sentence = " ".join(filtered_tokens)

    return filtered_tokens, clean_sentence
```

```python
test = "I absolutely looooved this product 😍😍 100%!!! Visit https://test.com"

tokens, sentence = preprocess_text(test)

print("Tokens:", tokens)
print("Clean Sentence:", sentence)
```

```python
#Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    results.append((sentence, tokens, clean_sentence))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Clean Sentence:", r[2])
```

```python
#Task 4: Token Analytics
import numpy as np

analytics = []

for sentence, tokens, clean_sentence in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    analytics.append((sentence, total_tokens, unique_tokens, avg_length))

for a in analytics:
    print("\nSentence:", a[0])
    print("Total Tokens:", a[1])
    print("Unique Tokens:", a[2])
    print("Avg Token Length:", round(a[3], 2))
```

```python
#Task 5: Frequency Analysis
from collections import Counter

all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
#Task 6: Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example usage
pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```
